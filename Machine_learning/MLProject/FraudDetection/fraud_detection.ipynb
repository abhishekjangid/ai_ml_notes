```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (classification_report, confusion_matrix,
                             precision_recall_curve, roc_auc_score, auc, f1_score,
                             precision_score, recall_score)
```

```python
# 1. Features and Labels
X = df.drop(columns=["Class", "Time"])
y = df["Class"]

# 2. Scaling 'Amount'
scaler = StandardScaler()
X["Amount"] = scaler.fit_transform(X[["Amount"]])

# 3. Train-Test Split (Stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

X_train.shape
```

```python
# 4. Model Training (weighted)
model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)

# 5. Predictions
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

# 6. Metrics
precision, recall, _ = precision_recall_curve(y_test, y_prob)
pr_auc = auc(recall, precision)
roc_auc = roc_auc_score(y_test, y_prob)
f1 = f1_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)

# 7. Report
print("=== Evaluation Metrics ===")
print(f"AUPRC (Precision-Recall AUC): {pr_auc:.4f}")
print(f"ROC-AUC: {roc_auc:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall: {rec:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    classification_report, confusion_matrix, precision_recall_curve,
    roc_auc_score, auc, f1_score, precision_score, recall_score
)
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Step 1: Load and Select Important Features
important_features = [
    'V11', 'V4', 'V2', 'V19', 'V8', 'V21', 'V27', 'V20', 'V28',
    'V17', 'V14', 'V12', 'V10', 'V16', 'V7', 'V3', 'V18', 'V1',
    'Amount'  # Keep amount as it's meaningful and not PCA'd
]

X = df[important_features]
y = df["Class"]

# Step 2: Scaling 'Amount'
scaler = StandardScaler()
X["Amount"] = scaler.fit_transform(X[["Amount"]])

# Step 3: Train-Test Split (Stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Step 4: Train Models
models = {
    "Random Forest": RandomForestClassifier(class_weight='balanced', random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

# Step 5: Evaluate Models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Metrics
    precision, recall, _ = precision_recall_curve(y_test, y_prob)
    pr_auc = auc(recall, precision)
    roc_auc = roc_auc_score(y_test, y_prob)
    f1 = f1_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)

    # Report
    print(f"\n=== Evaluation Metrics: {name} ===")
    print(f"AUPRC (Precision-Recall AUC): {pr_auc:.4f}")
    print(f"ROC-AUC: {roc_auc:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
```

### Inferencing

```python
import os
def load_fraud_detection_model(model_name='fraud_detection_rf', models_dir='model'):
    """
    Load a trained fraud detection model and preprocessing components

    Args:
        model_name (str): Name of the model to load ('RandomForest', 'LogisticRegression', 'XGBoost', 'SVM', 'KNN')
        models_dir (str): Directory containing the saved models

    Returns:
        dict: Dictionary containing model, scaler, features, and metadata
    """
    try:
        # Load model
        model_path = os.path.join(models_dir, f'{model_name}_model.pkl')
        model = joblib.load(model_path)

        # Load preprocessing components
        scaler_path = os.path.join(models_dir, 'amount_scaler.pkl')
        scaler = joblib.load(scaler_path)

        features_path = os.path.join(models_dir, 'feature_names.pkl')
        feature_names = joblib.load(features_path)

        # Load model info if available
        try:
            info_path = os.path.join(models_dir, f'model_info.pkl')
            model_info = joblib.load(info_path)
        except:
            model_info = None

        return {
            'model': model,
            'scaler': scaler,
            'feature_names': feature_names,
            'model_info': model_info,
            'model_name': model_name
        }

    except FileNotFoundError as e:
        raise FileNotFoundError(f"Model files not found. Please ensure models are trained and saved first. Error: {e}")

def preprocess_transaction_data(data, scaler, feature_names):
    """
    Preprocess transaction data for prediction

    Args:
        data (dict or pd.DataFrame): Transaction data
        scaler: Fitted StandardScaler for Amount column
        feature_names (list): List of required feature names

    Returns:
        pd.DataFrame: Preprocessed data ready for prediction
    """
    # Convert to DataFrame if needed
    if isinstance(data, dict):
        data = pd.DataFrame([data])
    elif isinstance(data, list):
        data = pd.DataFrame(data)

    # Check if all required features are present
    missing_features = set(feature_names) - set(data.columns)
    if missing_features:
        raise ValueError(f"Missing required features: {missing_features}")

    # Select and order features correctly
    data_processed = data[feature_names].copy()

    # Scale the Amount column
    data_processed['Amount'] = scaler.transform(data_processed[['Amount']])

    return data_processed


def predict_fraud(data, model_name='fraud_detection_rf', models_dir='model', return_probabilities=True):
    """
    Predict fraud for transaction data

    Args:
        data (dict, list, or pd.DataFrame): Transaction data to predict
        model_name (str): Name of the model to use
        models_dir (str): Directory containing saved models
        return_probabilities (bool): Whether to return probability scores

    Returns:
        dict or list: Prediction results
    """
    # Load model components
    model_components = load_fraud_detection_model(model_name, models_dir)

    # Preprocess data
    data_processed = preprocess_transaction_data(
        data,
        model_components['scaler'],
        model_components['feature_names']
    )

    # Make predictions
    predictions = model_components['model'].predict(data_processed)

    if return_probabilities:
        probabilities = model_components['model'].predict_proba(data_processed)

    # Format results
    results = []
    for i in range(len(data_processed)):
        result = {
            'prediction': int(predictions[i]),
            'prediction_label': 'FRAUD' if predictions[i] == 1 else 'NORMAL',
            'model_used': model_name
        }

        if return_probabilities:
            result.update({
                'fraud_probability': float(probabilities[i][1]),
                'normal_probability': float(probabilities[i][0]),
                'confidence': float(max(probabilities[i]))
            })

            # Risk assessment
            fraud_prob = probabilities[i][1]
            if fraud_prob > 0.7:
                risk_level = "HIGH"
            elif fraud_prob > 0.3:
                risk_level = "MEDIUM"
            else:
                risk_level = "LOW"
            result['risk_level'] = risk_level

        results.append(result)

    # Return single result if input was single transaction
    if len(results) == 1 and isinstance(data, dict):
        return results[0]

    return results


def create_sample_transaction():
    """Create a sample normal transaction for testing"""
    return {
        'V1': -1.359807134,
        'V2': -0.072781173,
        'V3': 2.536346738,
        'V4': 1.378155224,
        'V7': -0.073403395,
        'V8': 0.024330112,
        'V10': -0.338262553,
        'V11': 0.462388478,
        'V12': -0.168037598,
        'V14': -0.057613549,
        'V16': -0.270532677,
        'V17': -0.154195737,
        'V18': -0.097200297,
        'V19': 0.003286354,
        'V20': -0.012100313,
        'V21': -0.008142725,
        'V27': 0.066833982,
        'V28': 0.014207227,
        'Amount': 149.62
    }

def create_suspicious_transaction():
    """Create a sample suspicious transaction for testing"""
    return {
        'V1': 2.0,
        'V2': 1.5,
        'V3': -3.0,
        'V4': 2.5,
        'V7': -2.0,
        'V8': 1.8,
        'V10': -2.5,
        'V11': 3.0,
        'V12': -1.5,
        'V14': -4.0,
        'V16': 2.0,
        'V17': -3.5,
        'V18': 1.0,
        'V19': 2.2,
        'V20': -1.8,
        'V21': 0.5,
        'V27': -2.0,
        'V28': 0.8,
        'Amount': 5000.0
    }

def test_inference():
    """Test the inference functions with sample data"""
    print("=== FRAUD DETECTION INFERENCE TESTING ===\n")

    try:
        # Test single prediction
        print("1. Testing single transaction prediction:")
        normal_transaction = create_sample_transaction()
        result = predict_fraud(normal_transaction)

        print(f"   Transaction Type: Normal")
        print(f"   Prediction: {result['prediction_label']}")
        print(f"   Fraud Probability: {result['fraud_probability']:.4f}")
        print(f"   Risk Level: {result['risk_level']}")
        print(f"   Model Used: {result['model_used']}\n")


        print("2. Testing suspicious transaction:")
        suspicious_transaction = create_suspicious_transaction()
        result = predict_fraud(suspicious_transaction)

        print(f"   Transaction Type: Suspicious")
        print(f"   Prediction: {result['prediction_label']}")
        print(f"   Fraud Probability: {result['fraud_probability']:.4f}")
        print(f"   Risk Level: {result['risk_level']}")
        print(f"   Model Used: {result['model_used']}\n")


    except Exception as e:
        print(f"❌ Error during testing: {e}")
        print("Make sure models are trained and saved first by running 'model_training_and_saving.py'")






test_inference()
```
